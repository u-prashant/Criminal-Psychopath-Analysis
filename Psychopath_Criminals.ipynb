```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading dataset into Dataframe
criminals = pd.read_csv("psycho.csv")
```

```python
# Shape of the data i.e. data-points and features
print (criminals.shape)
```

```python
# Column Information
print(criminals.columns)
```

# Univariate Analysis

## HistoGram, PDF

```python
sns.FacetGrid(criminals, hue="Psycho ", size=5) \
   .map(sns.distplot, "I") \
   .add_legend()
```

```python
sns.FacetGrid(criminals, hue="Psycho ", size=5) \
   .map(sns.distplot, "We") \
   .add_legend()
```

```python
sns.FacetGrid(criminals, hue="Psycho ", size=5) \
   .map(sns.distplot, "was") \
   .add_legend()
```

```python
sns.FacetGrid(criminals, hue="Psycho ", size=5) \
   .map(sns.distplot, "went") \
   .add_legend()
```

```python
sns.FacetGrid(criminals, hue="Psycho ", size=5) \
   .map(sns.distplot, "had") \
   .add_legend()
```

```python
sns.FacetGrid(criminals, hue="Psycho ", size=5) \
   .map(sns.distplot, "because ") \
   .add_legend()
```

### Observations

1. Above PDFs shows that Psycho criminals uses more **I** than non-psychopath criminals.
2. **went** and **was** PDFs shows that non-pyschopath uses less was and went i.e. they use less past tense. They generally tend to talk in present tense.

## Violin Plot

```python
sns.violinplot(x="Psycho ", y="I", data=criminals, size=10)
plt.show()
```

### Observation

1. According to violin plot of **I**, non-psychopaths uses Personal pronoun less frequently than psychopaths.

# Bivariate Analysis

## Scatter Plot

```python
sns.set_style('whitegrid')
sns.FacetGrid(criminals, hue="Psycho ", size=5) \
   .map(plt.scatter, "I", "We") \
   .add_legend()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(criminals, hue="Psycho ", size=5) \
   .map(plt.scatter, "went", "was") \
   .add_legend()
```

## Pair Plot

```python
sns.set_style("whitegrid")
sns.pairplot(criminals, hue="Psycho ", vars=['I','We','was', 'went','had','because '], size=2)
plt.show()
```

### Observations

1. **I** is the best feature that can distinguish pyschopaths and non-psychopaths easily.
2. Features like **was, went, had** denoting past tense are not *overlapping* much and form separate clusters. Hence these features can also be used to classify Pyschopath criminals.
